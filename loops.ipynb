# while loops
a while loop in python is a control structre that repeats a block of code as longs as conditions is not respected or true
syntax:
while condition:
code to execute

condition:something that evalute  true or false the code inside the loops keeps runing while the condition is true 


```python
count=1
while count<=5:
    print(count)
    count+=1
```

```python
num=1
while num<20:
    print(num)
    num=num+1
```
