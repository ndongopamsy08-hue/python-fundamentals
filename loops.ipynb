# while loops
a while loop in python is a control structre that repeats a block of code as longs as conditions is not respected or true
syntax:
while condition:
code to execute

condition:something that evalute  true or false the code inside the loops keeps runing while the condition is true 


```python
count=1
while count<=5:
    print(count)
    count+=1
```

# for loop

a for loop in python is used to repeat a block of code for each item in a sequence e.g
- numbers
- list
- a string
- a range of numbers   

so a for loop run a fixed number of times we usuallly used a for looops where you you the exact number of iteriations to do

syntax
for varaibles in sequence:  
#code to be executed 

```python
for i in range(20):
    print (i)
    
```

```python
for i in range(0,22,2):
    print (i)
```

```python
#in a for loop using range(), the first value represents the starting point, the second value represents the stopping point (not included last number (stop) is NOT part of the result.), and the third value represents the step (increment or decrement).
for i in range(10,0,-1):
    print (i)
```

```python
for i in range(3):        # outer loop
    for j in range(2):    # inner loop
        print(i, j)
```

```python
for i in range(20,2,-1):
    print(i)
    
```

```python
for i in range(1, 6):        # controls rows (1 to 5)
    for j in range(1, i+1):  # prints from 1 to i
        print(j,end="")
    print()  # move to next line
```

```python
for i in range(1000,1010):
    print(f"PID:{i}")
```
