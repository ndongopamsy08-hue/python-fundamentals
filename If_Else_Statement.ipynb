# If_Else_statement
An if-else statement is a conditional statement where, if the first condition is not met, the code inside the else block will run instead.

```python
# A small exercise testing for eligibility 
age=int(input("enter your age:"))

if age >= 18:
    print("you are aligible to vote")
elif age < 0:
    print("YOU HAVE  NOT EVEN BEEN BORN YET STAY CALM")
elif age >= 100:
    print("YOU ARE TOO OLD TO VOTE ")
else:
    print("YOU MUST BE 18 AND ABOVE to vote  ")
    

```

```python
# a user is going to select an arthmetic operation either addition subtraction mult and division
num1=int(input("enter any num1 value of your choice: "))
num2=int(input("enter any num2 value of your choice: "))
print("1)--addition\n\n")
print("2)--subtractionn\n")
print("3)--multiplictionn\n")
print("4)--divisionn\n")
choose=int(input("choose which operation you want to offer"))
if choose==1:
    result=num1+num2
    print(result)
elif choose==2:
    result=num1-num2
    print(result)
elif choose==3:
    result=num1*num2
    print(result)
elif choose == 4:
    result = num1 / num2
    print(result)
else:
    print("invalid operation")
```
